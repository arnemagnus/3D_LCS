```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x, y, z = np.load('x.npy'), np.load('y.npy'), np.load('z.npy')
```

```python
lm3 = np.load('lm3.npy')
```

```python
grid = np.array(np.meshgrid(x,y,z,indexing='ij'))
```

```python
fig = plt.figure(figsize=(10,6),dpi=100)
ax = fig.add_subplot(111)
msh = ax.pcolormesh(x,y,np.log(lm3[...,-1]),cmap='viridis')
plt.colorbar(msh)
```
